a) El m√©todo de la secante (use los extremos como las estimaciones iniciales)


```python
def f(x):
    return 230*x**4+18*x**3+9*x**2-221*x-9

resultadoSecante9=newton(f,x0=-1,x1=0,tol=10**-6)
print(f"Resultado cuando [-1,0]: {resultadoSecante9}")

resultadoSecante10=newton(f,x0=0,x1=1,tol=10**-6)
print(f"Resultado cuando [0,1]: {resultadoSecante10}")

```

b. El m√©todo de Newton (use el punto medio como estimaci√≥n inicial)

```python
def f(x):
    return 230*x**4+18*x**3+9*x**2-221*x-9
def fprim(x):
    return 920*x**3+54*x**2+18*x-221

resultadoNewton5=newton(f,x0=-0.5,fprime=fprim,tol=10**-6)
print(f"Resultado cuando el punto medio es -0.5: {resultadoNewton5}")

resultadoNewton6=newton(f,x0=0.5,fprime=fprim,tol=10**-6)
print(f"Resultado cuando el punto medio es 0.5: {resultadoNewton6}")
```
