Encuentre los valores de A que hacen que la siguiente matriz sea singular

```python
import sympy as sp

alpha = sp.symbols('alpha')

A = sp.Matrix([
    [1, -1, alpha],
    [2,  2, 1],
    [0, alpha, -sp.Rational(3, 2)]
])

detA = A.det()

print("Determinante:")
sp.pprint(detA)

valores_alpha = sp.solve(detA, alpha)

print("\nValores de alpha que hacen singular la matriz:")
print(valores_alpha)
```
