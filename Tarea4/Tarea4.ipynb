1) Utilice el m√©todo de bisecci√≥n para encontrar soluciones precisas dentro de $10^{-6}$ para la ecuaci√≥n:
   
$ x^3 - 7x^2 + 14x - 6 = 0 $


        f   : funci√≥n de la que se busca la ra√≠z
        a   : extremo izquierdo del intervalo
        b   : extremo derecho del intervalo
        TOL : tolerancia (criterio de convergencia)
        N   : m√°ximo n√∫mero de iteraciones

```python
def f(x):
    return x**3 - 7*x**2 + 14*x - 6

def biseccion(f, a, b, TOL, N=1000):
    FA = f(a)
    FB = f(b)

        
    if abs(FA) < TOL:
         return a
     
    if abs(FB) < TOL:
        return b
    
    if FA * FB > 0:
        print("La funci√≥n no tiene ra√≠ces o tiene m√∫ltiples ra√≠ces en el intervalo dado.")
        return None

    for i in range(1, N + 1):
        p = (a + b) / 2
        FP = f(p)

        if abs(b - a) / 2 < TOL or abs(FP) < TOL :
            return p
        
        if FA * FP > 0:
            a = p
            FA = FP
        else:
            b = p

    print(f"El m√©todo fracas√≥ despu√©s de {N} iteraciones.")
    return None
resultado_A=biseccion(f,0,1,10**-2)
resultado_B=biseccion(f,1,3.2,10**-2)
resultado_C=biseccion(f,3.2,4,10**-2)

print(f"Raiz en [0,1] {resultado_A}")
print(f"Raiz en [1,3.2] {resultado_B}")
print(f"Raiz en [3.2,4] {resultado_C}")
```
